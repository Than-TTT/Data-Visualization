```python
import pandas as pd

# Load the datasets
high_pop_df = pd.read_csv('high_popularity_spotify_data.csv')
low_pop_df = pd.read_csv('low_popularity_spotify_data.csv')

# Inspect the datasets
print("High Popularity Data:")
print(high_pop_df.info())
print(high_pop_df.head())

print("\nLow Popularity Data:")
print(low_pop_df.info())
print(low_pop_df.head())
```

```python
# Concatenate the dataframes
merged_df = pd.concat([high_pop_df, low_pop_df], ignore_index=True)

# Inspect the merged dataframe
print("Merged Data:")
print(merged_df.info())
print(merged_df.head())

# Save the merged dataframe to a CSV file
merged_df.to_csv('merged_spotify_data.csv', index=False)
```
